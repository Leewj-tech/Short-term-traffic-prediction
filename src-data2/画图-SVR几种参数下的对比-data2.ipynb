```python
import matplotlib.pyplot as plt
from scipy.io import loadmat
import numpy as np
from math import sqrt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def quota(inv_y, inv_yhat,name):
    rmse_ = sqrt(mean_squared_error(inv_y, inv_yhat))
    mae_=mean_absolute_error(inv_y, inv_yhat)
    r2_ = r2_score(inv_y, inv_yhat)
    print(name,'的rmse:', rmse_, ' mae:', mae_, ' R2:', r2_)
```

```python
data0=loadmat('结果/svr1_result.mat')['true']
data1=loadmat('结果/svr1_result.mat')['pred']
data2=loadmat('结果/svr2_result.mat')['pred']
data3=loadmat('结果/svr3_result.mat')['pred']
```

```python
print(data0.shape,data1.shape,data2.shape,data3.shape)
```

```python
quota(data0,data1,'ELM')
quota(data0,data2,'ELM2')
quota(data0,data3,'ELM3')
```

```python
# In[7] 画图
plt.figure(figsize=(12,6))
plt.plot(data0,'-',label='real')
plt.plot(data1,'-',label='SVR')
plt.plot(data2,'-',label='SVR2')
plt.plot(data3,'-',label='SVR3')
plt.xlim((0,100))
plt.grid()
plt.legend()
plt.xlabel('Time/h',fontdict={"family":"Times New Roman"})
plt.ylabel('Congestion',fontdict={"family":"Times New Roman"})
plt.savefig('figure/SVR几种参数下预测结果图.png')
plt.show()
```
